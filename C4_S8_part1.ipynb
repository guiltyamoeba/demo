```python
#Check data for the extracted columns and convert them into the correct data types for further analysis.
import pandas as pd

data = pd.read_csv('C:/Users/91918/OneDrive/Desktop/SUDHEERA/DataScience with Gen Ai/Course4-Numpy python/Sprint8//DS1_C4_S8_Tendulkar_ODI_Data_Practice.csv')

# Extract the desired columns
extracted_data = data[['Runs', 'BF', '4s', '6s']]

# Drop rows with null values in any of the specified columns
extracted_data = extracted_data.dropna(subset=['Runs', 'BF', '4s', '6s'])

# Check the current data types
print(extracted_data.dtypes)

# Convert data types as needed
extracted_data['Runs'] = extracted_data['Runs'].astype(float)
extracted_data['BF'] = extracted_data['BF'].astype(int)
extracted_data['4s'] = extracted_data['4s'].astype(int)
extracted_data['6s'] = extracted_data['6s'].astype(int)

print(extracted_data.dtypes)
```

```python
#After importing the required libraries, calculate mean, median, and mode scores for the extracted columns: Runs, BF, 4s, and 6s.
import pandas as pd
import numpy as np

mean_scores = extracted_data[['Runs', 'BF', '4s', '6s']].mean()
median_scores = extracted_data[['Runs', 'BF', '4s', '6s']].median()
mode_scores = extracted_data[['Runs', 'BF', '4s', '6s']].mode().iloc[0]  # Assuming unique mode

# Print the results
print("Mean Scores:")
print(mean_scores)
print("\nMedian Scores:")
print(median_scores)
print("\nMode Scores:")
print(mode_scores)
```

```python
#Calculate the variance and standard deviation for the
extracted columns: Runs, BF, 4s, and 6s
import pandas as pd
import numpy as np


variance_scores = extracted_data[['Runs', 'BF', '4s', '6s']].var()
std_deviation_scores = extracted_data[['Runs', 'BF', '4s', '6s']].std()

# Print the results
print("Variance Scores:")
print(variance_scores)
print("\nStandard Deviation Scores:")
print(std_deviation_scores)
```
